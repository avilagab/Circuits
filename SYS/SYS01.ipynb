## Clase 01 - Introducción a Señales y Sistemas
### ¿Qué es una señal?
Una señal es la representación matemática de una o más variables independientes, en términos de una función. Algunos ejemplos de señales, en el contexto de la Ingeniería en Telecomunicaciones, pueden ser la corriente o el voltaje en el tiempo (En este caso, el tiempo es la variable independiente).
Existen dos tipos de señales, señales continuas o señales discretas.
### Señal continua
Representan una serie de datos que describen un fenómeno dado a lo largo de una variable independiente que es continua. Cuando la variable independiente es el tiempo **t**, se conocen como "Señales continuas en el tiempo". Es el caso de la temperatura **T** que sentimos en un día cualquiera **T(t)**, la cual varía lentamente y de forma continua. De igual manera, la voz **v** que transmitimos al hablar tiene una forma continua en el tiempo **v(t)**.
Sin embargo, la variable independiente no siempre es el tiempo. Por ejemplo, en el caso de un sonido que se emite a un volumen y frecuencias fijas, el nivel de volumen percibido **v** puede depender de la distancia **d** de quien lo escucha **v(d)**.

```python
#CLASE 01 - INTRODUCCIÓN A LAS SEÑALES
#Para representar una magnitud analógica
#El primer paso corresponde en importar las librerías matplotlib y numpy
import matplotlib.pyplot as plt
import numpy as np
#Posteriormente, se definen las características de la gráfica a dibujar
plt.rcParams["figure.figsize"] = [7, 3]
plt.rcParams["figure.autolayout"] = True
```

```python
#La variable independiente t se tomará en el rango [0,10] y se graficará cada 0.01 segundos.
t = np.arange(0, 10, 0.01) 

# Señal a graficar x(t) = Sin(PI*t)*e^(-0.5*t)
x = np.sin( np.pi * t) * np.exp (-0.5 * t)
fig, axs = plt.subplots()
axs.set_title("Señal continua en el tiempo")
axs.plot(t, x, color='C0', linewidth=2, label='x(t) = Sin(PI*t)*e^(-0.5*t)')
axs.set_xlabel("Tiempo")
axs.set_ylabel("Amplitud")
plt.legend(loc='upper right')
plt.show()
```

### Señal discreta
Representan una serie de datos que no son continuos y que, mediante un proceso de muestreo, describen un fenómeno dado a ciertos intervalos. En este caso, la señal es muestreda cada cierto tiempo t (intervalo de muestreo) y obtiene valores específicos. Si la señal discreta puede tomar sólo algunos valores específicos, se dice que es una señal digital.

```python
dt = 0.2 # Intervalo de muestreo
Fs = 1 / dt # frecuencia de muestreo
t = np.arange(0, 10, dt)

# Señal a graficar x(t) = Sin(PI*t)*e^(-0.5*t)
x = np.sin(np.pi * t) * np.exp (-0.5 * t)
fig, axs = plt.subplots()

axs.set_title("Señal discreta en el tiempo")
markerline, stemlines, baseline = plt.stem(t, x, label='x(t) = Sin(PI*t)*e^(-0.5*t)')
axs.set_xlabel("Tiempo")
axs.set_ylabel("Amplitud")
plt.legend(loc='upper right')
plt.show()
```
